## Предельные теоремы

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def print_experiments(cases):
    for i in range(len(cases)):
            print(str(i + 1) + '   ---   '+ str(cases[i]))
    
```

```python
def visualize_experiments(multi_cases, p, x_lim=1):
    for cases in multi_cases:
        n = len(cases)
        plt.xlim(0, n)
        plt.ylim(p - x_lim, p  + x_lim)
        means = np.empty(n)
        s = 0;
        for i in range(n):
            s += cases[i]
            means[i] = s / (i + 1)
        plt.plot(np.linspace(1, n, num=n, dtype=int), means)
    plt.grid()
    plt.xlabel('n')
    plt.ylabel('fn')
    plt.axhline(y = p, color = 'r', linestyle = '-')
    plt.show()
```

### 1. Теорема Бернулли

```python
def check_bernoulli_theorem(cases, p, eps, v):
    visualize_experiments([cases], p)
    fn = np.mean(cases)
    dif = abs(fn - p)
    print(f'Для {v} наблюдений получаем fn = {fn} => |{fn} - {p}| = {dif}')
    if (dif < eps) :
        print(f'{dif} < {eps} => для {v} наблюдений и при eps = {eps} теорема Бернулли выполняется\n')
    
```

```python
n = 1850
p = 0.5
cases = np.trunc(np.random.uniform(0, 1, size=n) + 0.5)
print_experiments(cases)

v1 = 170
v2 = n
eps1 = 0.1
eps2 = 0.03

check_bernoulli_theorem(cases[:v1], p, eps1, v1)
check_bernoulli_theorem(cases[:v2], p, eps2, v2)
```

### 2. Закон больших чисел в форме Чебышева

#### Испытание практически достоверных событий. 

#### Равномерное распределение распределение чисел на отрезке [0, 1]

```python
def check_chebyshev_lln(cases, M, eps, v):
    visualize_experiments([cases], M)
    fn = np.mean(cases)
    dif = abs(fn - M)
    print(f'Для {v} наблюдений получаем fn = {fn} => |{fn} - {M}| = {dif}')
    if (dif < eps) :
        print(f'{dif} < {eps} => для {v} наблюдений и при eps = {eps} неравенство выполняется\n')
```

```python
n = 1125
M = 0.5
cases = np.array(np.random.uniform(0, 1, size=n))
print_experiments(cases)
eps1 = 0.1
eps2 = 0.02
v1 = 45
v2 = n

check_chebyshev_lln(cases[:v1], M, eps1, v1)
check_chebyshev_lln(cases[:v2], M, eps2, v2)
```

#### Экспонециальное распределение с м.о. = 1

```python
eps1 = 0.2
eps2 = 0.05
M = 1

v1 = math.ceil(5.4 * M / (eps1 ** 2))
v2 =  math.ceil(5.4 * M / (eps2 ** 2))
cases = np.array([-np.log(x) for x in np.random.uniform(0, 1, size=v2)])
print_experiments(cases)

check_chebyshev_lln(cases[:v1], M, eps1, v1)
check_chebyshev_lln(cases[:v2], M, eps2, v2)
```

#### Невыполнение закона больших чисел. Распределение Коши.

```python
multi_cases = np.array([np.tan(np.random.uniform(low=0, high=np.pi, size=1000)) for i in range (7)])
means = list()
for cases in multi_cases:
    means.append(np.mean(cases))
    visualize_experiments([cases], 0, x_lim=20)
    
    
for m in means:
    print(f'Среднее значение выборки: {m}')
```

#### б) разброс средних (для равномерного распределения)

```python
ns = [10, 40, 160, 640]
k = 20
results = dict()

multi_cases = [np.array(np.random.uniform(0, 1, size=max(ns))) for _ in range(k)]
results = dict()

for n in ns:
    means = list()
    for cases in multi_cases:
        s_cases = cases[:n]
        means.append(np.mean(s_cases))
    results[n] = [min(means), max(means), np.std(means), means]

print('n\tмин. среднее\t\tмакс. среднее\t\t\tстанд. откл')
for n in ns:
    r = results[n]
    print(f'{n}\t{r[0]}\t{r[1]}\t\t{r[2]}')

plt.xlim(0.3, 0.7)
for n in ns:
    r = results[n]
    y = [n for _ in range(k)]
    plt.plot(r[3], y, marker=".")
plt.xlabel('std')
plt.ylabel('n')
plt.grid()
plt.show()
```

### 4. Основная теорема статистики

```python
ns = [10, 40, 160, 640, 2560]

for n in ns:
    cases = np.array(np.random.uniform(0, 1, size=n))
    sorted_cases = np.sort(cases)
    print(f'n = {n}')
    plt.plot(np.linspace(0, 1, num=2, dtype=int), np.linspace(0, 1, num=2, dtype=int))
    F = [i / n for i in range(n)]
    plt.step(sorted_cases, F)
    plt.grid()
    plt.show()
```
